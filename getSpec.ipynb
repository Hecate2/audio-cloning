```python
import wave
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import scipy.signal as signal
import scipy.io.wavfile
```

```python
f, d = wavfile.read('5.wav')
plt.figure(0)
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.title('Signal Plot')
plt.plot(d)
fs, t, Zxx= signal.stft(d, window='hamming', nperseg=512, noverlap=0.75*512)
# plt.plot(Zxx)
plt.figure(1)
plt.xlabel('Samples')
plt.ylabel('Frequency')
plt.title('Spectrogram')
plt.pcolormesh(t, fs, np.abs(Zxx), cmap='hot')
```
